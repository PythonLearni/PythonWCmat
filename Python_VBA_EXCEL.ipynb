```python
# -*- coding: utf-8 -*-
"""
Created on Mon Jun 11 10:14:09 2018

@author: n886528
"""
import pandas as pd
import os
import numpy as np



def compare_chart(workbook,sheetname,df,\
                  row_start,row_end,\
                  additional={}):
    chart = workbook.add_chart({'type': 'line'})
    colnames=df.columns.values.tolist()
    cat_col_index= colnames.index('Q')
    # start from 2
    color_convention = {i:color_schema[j] for i,j in zip(list(range(2,cat_col_index)),colnames[2:cat_col_index])}
    dash_set = [colnames.index(x) if 'CCAR' in x else 'Nothing' for x in colnames]
    #color_convention = {2:'black',3:'black',4:'#FF9900',5:'#FF9900',6:'green',7:'green',8:'red',9:'red'}
    for i in range(2,cat_col_index):
        chart.add_series({
            'categories': [sheetname,row_start,cat_col_index,row_end,cat_col_index],
            'values':     [sheetname,row_start,i,row_end,i],
            'name':      [sheetname,0,i],
            'line': {'color': color_convention[i],'dash_type': 'square_dot' if i in dash_set else 'solid'}
        })
    y_setting = {'major_gridlines': {'visible': False},
                      'minor_tick_type':'none',
                      'major_tick_type':'none'}
    y_setting.update(additional)
    chart.set_y_axis(y_setting)
    chart.set_legend({'position': 'bottom'})
    # Turn off chart legend. It is on by default in Excel.
    chart.set_legend({'position': 'bottom'})
    
    chart.set_x_axis({'position_axis': 'on_tick',
                      'minor_tick_type':'none',
                      'major_tick_type':'none'
                     })
    chart.set_title({
        'name': [sheetname,row_start,1]
    })
    return chart


def indep_chart(workbook,sheetname,df,\
                  row_start,row_end,\
                  additional={}):
    chart_indep = workbook.add_chart({'type': 'line'})
    colnames=df.columns.values.tolist()
    cat_col_index= colnames.index('Q')
    # start from 2
    color_convention = {i:color_schema[j] for i,j in zip(list(range(2,cat_col_index)),colnames[2:cat_col_index])}
    dash_set = [colnames.index(x) if 'CCAR' in x else 'Nothing' for x in colnames]

    #color_convention = {2:'black',3:'black',4:'#FF9900',5:'#FF9900',6:'green',7:'green',8:'red',9:'red'}
    for i in set(list(range(2,cat_col_index)))-set(dash_set): 
        chart_indep.add_series({
            'categories': [sheetname,row_start,cat_col_index,row_end,cat_col_index],
            'values':     [sheetname,row_start,i,row_end,i],
            'name':      [sheetname,0,i],
            'line': {'color': color_convention[i],'dash_type': 'square_dot' if i in dash_set else 'solid'}
        })
    
    # Configure the chart axes.
    y_setting = {'major_gridlines': {'visible': False},
                      'minor_tick_type':'none',
                      'major_tick_type':'none'}
    y_setting.update(additional)
    print(y_setting)
    chart_indep.set_y_axis(y_setting)
    chart_indep.set_legend({'position': 'bottom'})
    # Turn off chart legend. It is on by default in Excel.
    chart_indep.set_legend({'position': 'bottom'})
    
    chart_indep.set_x_axis({'position_axis': 'on_tick',
                      'minor_tick_type':'none',
                      'major_tick_type':'none'
                     })
    chart_indep.set_title({
        'name': [sheetname,row_start,1]
    })
    return chart_indep


def generate_excel_plot(data_file,sheetname_list,excel_file,color_schema,qrt_len=9):
    all_data = pd.read_excel(io = data_file,
              sheetname = sheetname_list
              )
    writer = pd.ExcelWriter(excel_file, engine='xlsxwriter')
    for i in sheetname_list:
        quarter_len = qrt_len
        df = all_data[i]
        if i in ['CNI','CRE']:
            adj_quarter_len =quarter_len + 1
            df['Q'] = ['Q'+str(i) for i in range(0,adj_quarter_len)]*len(np.unique(df['MacroVar']))
            start = 1
            num_macro = len(np.unique(df['MacroVar']))
            start_row = [start + adj_quarter_len*i  for i in range(num_macro)]
            end_row = [k+adj_quarter_len-1 for k in start_row]
        else:
            df['Q'] = ['Q'+str(i) for i in range(1,quarter_len+1)]*len(np.unique(df['MacroVar']))
            
            start = 1
            
            num_macro = len(np.unique(df['MacroVar']))
            start_row = [start + quarter_len*i  for i in range(num_macro)]
            end_row = [k+quarter_len-1 for k in start_row]
        position = [1 + 15*i for i in range(num_macro)]
        
        
        # Create a Pandas Excel writer using XlsxWriter as the engine.
        
        sheet_name = i
        
        
        df.to_excel(writer, sheet_name=sheet_name,index = None)
        
        # Access the XlsxWriter workbook and worksheet objects from the dataframe.
        # This is equivalent to the following using XlsxWriter on its own:
        #
        #    workbook = xlsxwriter.Workbook('filename.xlsx')
        #    worksheet = workbook.add_worksheet()
        workbook = writer.book
        worksheet = writer.sheets[sheet_name]
        
        for row_start,row_end,pos in zip(start_row,end_row,position):
            min_y = df.iloc[(row_start-1):row_end,2:(len(df.keys())-1)].min().min()
            max_y = df.iloc[(row_start-1):row_end,2:(len(df.keys())-1)].max().max()
            print(min_y,max_y)
            # Create a chart object.
            if min_y<0:
                add = {'crossing':round(min_y*11,1)/10 if abs(min_y)<1 else round(min_y*1.1,1),'min':round(min_y*11,1)/10 if abs(min_y)<1 else round(min_y*1.1,1),'max':round(max_y*11,1)/10 if abs(max_y)<1 else round(max_y*1.1,1)}
            else:
                add = {}
            worksheet.insert_chart('M'+str(pos), indep_chart(workbook=workbook,row_start=row_start,row_end=row_end,sheetname=sheet_name,additional=add,df=df))
            worksheet.insert_chart('U'+str(pos), compare_chart(workbook=workbook,row_start=row_start,row_end=row_end,sheetname=sheet_name,additional=add,df=df))
    
    
    # Close the Pandas Excel writer and output the Excel file.
    writer.save()
    writer.close()

    generate_excel_plot(data_file = data_file,
                    sheetname_list=sheetname_list,
                    excel_file=excel_file,
                    color_schema=color_schema,
                    qrt_len=9)


```
